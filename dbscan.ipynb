```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/gayathri/Desktop/data-mining/data4_subset.csv')
features = data.drop('Exited', axis=1)

# Standardize the features
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

silhouette_scores = []
num_clusters = []
min_samples_range = range(50, 701, 50)  # From 50 to 500, stepping by 50
eps_value = 2  

for min_samples in min_samples_range:
    dbscan = DBSCAN(eps=eps_value, min_samples=min_samples)
    labels = dbscan.fit_predict(features_scaled)
    
    unique_dbscan_values = np.unique(labels)
    num_clusters_formed = len(unique_dbscan_values) - 1 if -1 in unique_dbscan_values else len(unique_dbscan_values)
    num_clusters.append(num_clusters_formed)
    count_minus_1 = np.sum(labels == -1)
    
    print("eps={}, min_samples={}, num_clusters={}, num_outliers={}".format(eps_value, min_samples, num_clusters_formed, count_minus_1))
    
    # Ensure there's more than one cluster excluding noise
    if num_clusters_formed > 1:
        score = silhouette_score(features_scaled, labels)
        silhouette_scores.append(score)
    else:
        silhouette_scores.append(None)  # Append None if only one cluster is found

    print("eps={}, min_samples={}, silhouette_score={}".format(eps_value, min_samples, silhouette_scores[-1]))
```

```python
plt.figure(figsize=(5,5))
plt.plot(list(min_samples_range), silhouette_scores, marker='o')
plt.title('Silhouette Scores for Varying min_samples')
plt.xlabel('min_samples')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

plt.figure(figsize=(5,5))
plt.plot(list(min_samples_range), num_clusters, marker='o')
plt.title('Silhouette Scores for Varying min_samples')
plt.xlabel('min_samples')
plt.ylabel('Clusters')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

#eps - epsilon - Eps is the radius of the neighborhood around each point
#min_samples - minimum number of points that must be within the neighborhood of a point for it to be considered a core point.

# Apply DBSCAN
dbscan = DBSCAN(eps=2, min_samples=50)
dbscan_clusters = dbscan.fit_predict(features_scaled)

#Add Cluster Labels to DataFrame
data['DBSCAN_Cluster'] = dbscan_clusters

unique_dbscan_values = data['DBSCAN_Cluster'].unique()
print(unique_dbscan_values)    #number of clusters and -1 indicates outliers
count_minus_1 = (data['DBSCAN_Cluster'] == -1).sum()
print(count_minus_1) #number of outliers

from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(features_scaled, data['DBSCAN_Cluster'])
print("Average Silhouette Score:", silhouette_avg)

data.to_csv('/Users/gayathri/Desktop/data-mining/clustered_data.csv', index=False)
```

```python
from sklearn.manifold import TSNE

# Apply t-SNE to reduce dimensions to 2 for visualization
tsne = TSNE(n_components=2, random_state=42)
tsne_results = tsne.fit_transform(features_scaled)

# Scatter plot of the t-SNE results colored by cluster label
plt.figure(figsize=(10, 8))
scatter = plt.scatter(tsne_results[:, 0], tsne_results[:, 1], c=data['DBSCAN_Cluster'], cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('t-SNE of Clustered Data')
plt.xlabel('t-SNE Feature 1')
plt.ylabel('t-SNE Feature 2')
plt.colorbar(scatter)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Apply PCA to reduce dimensions to 2 for visualization
pca = PCA(n_components=2)
principal_components = pca.fit_transform(features_scaled)

# Scatter plot of the two principal components colored by cluster label
plt.figure(figsize=(10, 8))
scatter = plt.scatter(principal_components[:, 0], principal_components[:, 1], c=data['DBSCAN_Cluster'], cmap='viridis', edgecolor='k', alpha=0.7)
plt.title('PCA of Clustered Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(scatter)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Filter out outliers
clustered_data = data[data['DBSCAN_Cluster'] != -1]

# Calculate silhouette scores for each cluster separately
unique_clusters = clustered_data['DBSCAN_Cluster'].unique()
for cluster_label in unique_clusters:
    cluster_subset = clustered_data[clustered_data['DBSCAN_Cluster'] == cluster_label]
    if cluster_subset.shape[0] > 1:  # Silhouette score is not defined for a single sample
        silhouette_avg = silhouette_score(features_scaled[cluster_subset.index], cluster_subset['DBSCAN_Cluster'])
        print(f"Silhouette Score for cluster {cluster_label}: {silhouette_avg}")
```
